```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import numpy as np

test_folder_path = "DGE_test_cows_mixed"
cows_test = glob.glob(test_folder_path+"/*/*")
cows_test_class_names = []
for name in cows_test:
    if (name[-2:] != '_2' and name[-2:] != '_1') or name == "Cow_2" or name == "Cow_1":
        #print(name[-2:])
        class_cow = (name[len(test_folder_path+"\\"):len(test_folder_path+"\\")+1], name[len(test_folder_path+"\\5\\"):])
        cows_test_class_names.append(class_cow)
    else:
        print(name)
        print(name[-2:0])
        pass
#print(cows_test_class_names)

val_folder_path = "DGE_val_cows_mixed"
cows_val = glob.glob(val_folder_path+"/*/*")
cows_val_class_names = []
for name in cows_val:
    if (name[-2:] != '_2' and name[-2:] != '_1') or name == "Cow_2" or name == "Cow_1":
        #print(name[-2:])
        class_cow = (name[len(val_folder_path+"\\"):len(val_folder_path+"\\")+1], name[len(val_folder_path+"\\5\\"):])
        cows_val_class_names.append(class_cow)
    else:
        print(name)
        print(name[-2:0])
        pass
#print(cows_val_class_names)

training_folder_path = "DGE_training_cows_mixed"
cows_training = glob.glob(training_folder_path+"/*/*")
cows_training_class_names = []
for name in cows_training:
    if (name[-2:] != '_2' and name[-2:] != '_1') or name == "Cow_2" or name == "Cow_1":
        #print(name[-2:])
        class_cow = (name[len(training_folder_path+"\\"):len(training_folder_path+"\\")+1], name[len(training_folder_path+"\\5\\"):])
        cows_training_class_names.append(class_cow)
    else:
        print(name)
        print(name[-2:0])
        pass
#print(cows_training_class_names)

total_names = cows_training_class_names + cows_test_class_names + cows_val_class_names
#print(total_names[0][0])

count_train = []
for element in cows_training_class_names:
    count_train.append(int(element[0][0]))

count_val = []
for element in cows_val_class_names:
    count_val.append(int(element[0][0]))

count_test = []
for element in cows_test_class_names:
    count_test.append(int(element[0][0]))


count_train_hist, bins = np.histogram(count_train, [2, 3, 4, 5, 6, 7])
print('train', count_train_hist, bins)
count_val_hist, bins = np.histogram(count_val, [2, 3, 4, 5, 6, 7])
print('val', count_val_hist, bins)
count_test_hist, bins = np.histogram(count_test, [2, 3, 4, 5, 6, 7])
print('test', count_test_hist, bins)

total_count_hist = np.vstack((np.vstack((count_train_hist, count_val_hist)), count_test_hist))

plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(10,10))
plt.bar([2, 3, 4, 5, 6], count_train_hist, width=0.5)
plt.bar([2, 3, 4, 5, 6], count_val_hist, width=0.5, bottom=count_train_hist)
plt.bar([2, 3, 4, 5, 6], count_test_hist, width=0.5, bottom=count_train_hist+count_val_hist)
plt.legend(['Training', 'Validation', 'Testing'])
plt.xlabel('Class')
plt.ylabel('Number of Cows')
plt.title('Data Set Split by Cows')
plt.savefig('Figures/Data_set_split_num_cows.png', format='png')
```

```python
#makes a few plots to display the NRMSE of the images in the data set from frame to frame

import numpy as np
import imageio
import matplotlib.pyplot as plt
import glob
import skimage.metrics
plt.rcParams.update({'font.size': 15})

# img1 = imageio.imread("DGE_test\\2\\0_DGE.tif")
# img2 = imageio.imread("DGE_test\\2\\100_DGE.tif")
# img3 = img1-img2
# channel = 0

# plt.figure(figsize=(10,10))
# plt.imshow((img3)[:, :, channel]) 
# print(np.sum(img3[:, :, channel]), np.min(img3[:, :, channel]), np.max(img3[:, :, channel]))

#filenames = glob.glob("DGE_test\\2\\*")
total_sum = []
temp_SSIM = []
total_SSIM = []

for bcs in range(2, 7):
    cows = glob.glob("DGE_test_cows_oct\\"+str(bcs)+"\\*")
    #print(cows)
    for cow in cows:
        temp_SSIM = []
        filenames = glob.glob(cow+"\\*")
        img_prev = imageio.imread(filenames[0])
        #print(filenames[0], 'img_true')
        for image in filenames:
            #img_prev = img_test
            #print(filenames[0], image)
            img_test = imageio.imread(image)
            #temp_NRMSE.append(np.mean(np.square(img_test - img_prev)))
            temp_SSIM.append(skimage.metrics.structural_similarity(img_prev, img_test, multichannel=True))
            img_prev = img_test

        total_SSIM.append(temp_SSIM)

print(np.shape(total_SSIM)[0])

plt.figure(figsize=(10,10))
#plt.subplot(1, 2, 1)
for i in range(0, np.shape(total_SSIM)[0], 4):
    try:
        plt.scatter(np.arange(0, 30), total_SSIM[i][0:30])
    except:
        pass
plt.title('SSIM Frame to Frame (Cattle Held in Lane)')
plt.xlabel("Frame Number")
plt.ylabel("SSIM")
plt.ylim(0.75, 1)

mean_oct = []
for i in range(0, np.shape(total_SSIM)[0]):
    mean_oct.append(np.mean(total_SSIM[i]))


# #np.shape(total_NRMSE[i])[0]
# plt.figure()
# plt.plot(np.arange(0, np.shape(total_NRMSE[0])[0]), total_NRMSE[0])
# plt.figure()
# plt.plot(np.arange(0, np.shape(total_NRMSE[7])[0]), total_NRMSE[7])

total_sum = []
temp_SSIM = []
total_SSIM = []

for bcs in range(2, 7):
    cows = glob.glob("DGE_training_total_cows\\"+str(bcs)+"\\*")
    #print(cows)
    for cow in cows:
        temp_SSIM = []
        filenames = glob.glob(cow+"\\*")
        img_prev = imageio.imread(filenames[0])
        #print(filenames[0], 'img_true')
        for image in filenames:
            #img_prev = img_test
            #print(filenames[0], image)
            img_test = imageio.imread(image)
            #temp_NRMSE.append(np.mean(np.square(img_test - img_prev)))
            temp_SSIM.append(skimage.metrics.structural_similarity(img_prev, img_test, multichannel=True))
            img_prev = img_test

        total_SSIM.append(temp_SSIM)
print(np.shape(total_SSIM))
plt.figure(figsize=(10,10))
#plt.subplot(1, 2, 2)
for i in range(0, np.shape(total_SSIM)[0]):
    try:
        plt.scatter(np.arange(1, 5, 1), total_SSIM[i][1:5])
    except:
        pass
plt.title('SSIM Frame to Frame (Cattle Walked Through Lane)')
plt.xlabel("Frame Number")
plt.ylabel("SSIM")
plt.ylim(0.75, 1)

mean_train = []
for i in range(0, np.shape(total_SSIM)[0]):
    mean_train.append(np.mean(total_SSIM[i]))

print(np.mean(mean_oct))
print(np.mean(mean_train))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import imageio
import glob
import skimage.transform
import skimage.metrics
plt.rcParams.update({'font.size': 15})

okapi = glob.glob('101_ObjectCategories\\okapi\\*')
img1 = skimage.transform.resize(imageio.imread(okapi[0]), (224, 224, 3))
okapi_NRMSE = []
for image in okapi[1:]:
    img_test = skimage.transform.resize(imageio.imread(image), (224, 224, 3))
    #okapi_NRMSE.append(np.mean(np.square(img_test - img1)))
    okapi_NRMSE.append(skimage.metrics.normalized_root_mse(img1, img_test, normalization='mean'))

faces = glob.glob('101_ObjectCategories\\faces\\*')
img1 = skimage.transform.resize(imageio.imread(faces[0]), (224, 224, 3))
faces_NRMSE = []
for image in faces[1:]:
    img_test = skimage.transform.resize(imageio.imread(image), (224, 224, 3))
    #faces_NRMSE.append(np.mean(np.square(img_test - img1)))
    faces_NRMSE.append(skimage.metrics.normalized_root_mse(img1, img_test, normalization='mean'))

faces_easy = glob.glob('101_ObjectCategories\\faces_easy\\*')
img1 = skimage.transform.resize(imageio.imread(faces_easy[0]), (224, 224, 3))
faces_easy_NRMSE = []
for image in faces_easy[1:]:
    img_test = skimage.transform.resize(imageio.imread(image), (224, 224, 3))
    #faces_easy_NRMSE.append(np.mean(np.square(img_test - img1)))
    faces_easy_NRMSE.append(skimage.metrics.normalized_root_mse(img1, img_test, normalization='mean'))


dalmatians = glob.glob('101_ObjectCategories\\dalmatian\\*')
img1 = skimage.transform.resize(imageio.imread(dalmatians[0]), (224, 224, 3))
dalmatians_NRMSE = []
for image in dalmatians[1:]:
    img_test = skimage.transform.resize(imageio.imread(image), (224, 224, 3))
    #faces_easy_NRMSE.append(np.mean(np.square(img_test - img1)))
    dalmatians_NRMSE.append(skimage.metrics.normalized_root_mse(img1, img_test, normalization='mean'))

dolphin = glob.glob('101_ObjectCategories\\dolphin\\*')
img1 = skimage.transform.resize(imageio.imread(dolphin[0]), (224, 224, 3))
dolphin_NRMSE = []
for image in dolphin[1:]:
    img_test = skimage.transform.resize(imageio.imread(image), (224, 224, 3))
    #faces_easy_NRMSE.append(np.mean(np.square(img_test - img1)))
    dolphin_NRMSE.append(skimage.metrics.normalized_root_mse(img1, img_test, normalization='mean'))


# faces_okapi_NRMSE = []
# for okapi_img, face_img in zip(okapi[:50], faces[:50]):
#     okapi_test = skimage.transform.resize(imageio.imread(okapi_img), (224, 224, 3))
#     face_test = skimage.transform.resize(imageio.imread(face_img), (224, 224, 3))
#     #faces_okapi_NRMSE.append(np.mean(np.square(face_test - okapi_test)))
#     faces_okapi_NRMSE.append(skimage.metrics.normalized_root_mse(face_test, okapi_test, normalization='mean'))

# faces_faces_NRMSE = []
# for easy_img, face_img in zip(faces_easy[:50], faces[:50]):
#     easy_test = skimage.transform.resize(imageio.imread(easy_img), (224, 224, 3))
#     face_test = skimage.transform.resize(imageio.imread(face_img), (224, 224, 3))
#     #faces_faces_NRMSE.append(np.mean(np.square(easy_test - face_test)))
#     faces_faces_NRMSE.append(skimage.metrics.normalized_root_mse(face_test, easy_test, normalization='mean'))


plt.figure(figsize=(10,10))
plt.scatter(np.arange(0, 30), okapi_NRMSE[:30])
plt.scatter(np.arange(0, 30), faces_NRMSE[:30])
plt.scatter(np.arange(0, 30), faces_easy_NRMSE[:30])
#plt.scatter(np.arange(0, 30), faces_okapi_NRMSE[:30])
plt.scatter(np.arange(0, 30), dalmatians_NRMSE[:30])
plt.scatter(np.arange(0, 30), dolphin_NRMSE[:30])
#plt.scatter(np.arange(0, 30), faces_faces_NRMSE[:30])
plt.legend(['Okapi', 'Faces (hard)', 'Faces (easy)', 'Dalmation', 'Dolphin'])
plt.xlabel('Image Number')
plt.ylabel('NRMSE')
plt.title('NRMSE of CalTech101 Categories')
plt.ylim(0,2)

total_NRMSE_101 = np.vstack((okapi_NRMSE[:30], faces_NRMSE[:30]))
total_NRMSE_101 = np.vstack((total_NRMSE_101[:30], faces_easy_NRMSE[:30]))
total_NRMSE_101 = np.vstack((total_NRMSE_101[:30], dalmatians_NRMSE[:30]))
total_NRMSE_101 = np.vstack((total_NRMSE_101[:30], dolphin_NRMSE[:30]))
#total_NRMSE_101 = np.vstack((total_NRMSE_101[:30], faces_faces_NRMSE[:30]))
print(np.mean(total_NRMSE_101, axis=1))
```
